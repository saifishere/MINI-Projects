# Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df, test_size = 0.25, shuffle = True)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
df_train.head()
```

```python
x_train = df_train.drop(columns=['target', 'target_name']).values
y_train = df_train['target'].values
```

```python
x_test = df_test.drop(columns=['target', 'target_name']).values
y_test = df_test['target'].values
```

# Modelling Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

# Using Cross Validation

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
model = LogisticRegression(max_iter = 200)
accuracies = cross_val_score(model, x_train, y_train, cv=5, scoring='accuracy')
np.mean(accuracies)
```

```python
y_pred = cross_val_predict(model, x_train, y_train, cv=5)
```

```python
y_pred
```

```python
predicted_correctly = y_pred == y_train
```

```python
predicted_correctly
```

```python
not_predicted_correctly = ~predicted_correctly
x_train[not_predicted_correctly]
```

```python
df_predictions = df_train.copy()
```

```python
df_predictions['correct_predictions'] = predicted_correctly
```

```python
df_predictions['predictions'] = y_pred
```

```python
df_predictions['predictions_label'] = df_predictions['predictions'].map({0:"setosa", 1:"versicolor", 2:"virginica"})
```

```python
df_predictions.head()
```
